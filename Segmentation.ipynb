```python
columns = ['Price', 'Year', 'Kilometer', 'Fuel Type', 'Transmission', 'Drivetrain']
df = vehicle[columns].copy()
```

```python
df=df.dropna()
df
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
label_encoders = {}
for column in ['Fuel Type', 'Transmission', 'Drivetrain']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

label_encoders
```

```python
scaler = StandardScaler()
df[['Price', 'Year', 'Kilometer']] = scaler.fit_transform(df[['Price', 'Year', 'Kilometer']])
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 10))
for i, column in enumerate(df.columns[:-1], 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
inertia = []
silhouette_scores = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(df, kmeans.labels_))
```

```python
# Elbow Curve
plt.figure(figsize=(7, 5))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid()
plt.show()
```

```python
# Silhouette Score
plt.figure(figsize=(7, 5))
plt.plot(K, silhouette_scores, 'go-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Optimal k')
plt.show()
```

```python
optimal_k = 4 
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
df['Segment'] = kmeans.fit_predict(df)
df.head()
```

```python
sns.pairplot(df, hue='Segment', palette='viridis')
```
